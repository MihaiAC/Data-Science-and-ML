```python
# Select all the rows for which Pclass is 2 or 3
# class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
class_23 = titanic[titanic.Pclass.isin([2, 3])]
class_23.head(10)
```

```python
# Filter out the rowss for which age is not known
age_no_na = titanic[titanic["Age"].notna()]
age_no_na.head()
```

```python
# Filter on one column + select two other columns
adult_names = titanic.loc[titanic["Age"] > 35, ["Name", "Age"]]
adult_names.head()
```
